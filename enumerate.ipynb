```python
import numpy as np
import PIL
from PIL import Image 
import glob
```

```python
load_path1 = '/Users/jordankruguer/Desktop/neuro-140/140-data/processed-data/abstract_line/*.jpg'
save_path1 = '/Users/jordankruguer/Desktop/neuro-140/140-data/processed-data/abstract_line_enum/'

load_path2 = '/Users/jordankruguer/Desktop/neuro-140/140-data/processed-data/line_target/*.jpg'
save_path2 = '/Users/jordankruguer/Desktop/neuro-140/140-data/processed-data/line_target_enum/'


```

```python
images1 = []
images2 = []
```

```python
for filename in glob.glob(load_path1):
    print(filename)
    img = Image.open(filename)
    images1.append(img)
```

```python
for filename in glob.glob(load_path2):
    print(filename)
    img = Image.open(filename)
    images2.append(img)
```

```python
for [i,new] in enumerate(images1):
    new.save('{}{}{}'.format(save_path1,i+1,'.jpg'))
    
```

```python
for [i,new] in enumerate(images2):
    new.save('{}{}{}'.format(save_path2,i+1,'.jpg'))
```
