```python
load_path_abstract = '/Users/jordankruguer/Desktop/neuro-140/140-data/raw-data/abstract/*.jpg' 
load_path_line = '/Users/jordankruguer/Desktop/neuro-140/140-data/raw-data/line/*.jpg' 
load_path_target = '/Users/jordankruguer/Desktop/neuro-140/140-data/raw-data/target/*.jpg' 

save_path_1 = '/Users/jordankruguer/Desktop/neuro-140/140-data/processed-data/abstract_line/'
save_path_2 = '/Users/jordankruguer/Desktop/neuro-140/140-data/processed-data/line_target/'
```

```python
abstract = {}
line = {}
target = {}
```

```python
for filename in glob.glob(load_path_line):
    print(filename)
    img = Image.open(filename)
    filename = os.path.basename(filename)
    line[filename] = img
    
```

```python
for filename in glob.glob(load_path_target):
    print(filename)
    img = Image.open(filename)
    filename = os.path.basename(filename)
    target[filename] = img
    
```

```python
def get_concat_h(im1, im2):
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst
```

```python
print(target.keys())
```

```python
for filename in line.keys():
    img = get_concat_h(line[filename],target[filename])
    img.save('{}{}'.format(save_path_2,filename))
    
```
